```python
crop_dict={
    'rice':1,
    'maize':2,
    'chickpea':3,
    'kidneybeans':4,
    'pigeonpeas':5,
    'mothbeans':6,
    'mungbean':7,
    'blackgram':8,
    'lentil':9,
    'pomegranate':10,
    'banana':11, 
    'mango':12, 
    'grapes':13, 
    'watermelon':14,
    'muskmelon':15,
    'apple':16,
    'orange':17,
    'papaya':18,
    'coconut':19,
    'cotton':20,
    'jute':21, 
    'coffee':22

}
```

```python
df.label=df.label.map(crop_dict)
```

```python
reverse_crop_dict={v:k for k,v in crop_dict.items()}
```

```python
reverse_crop_dict
```

```python
df.head()
```

```python
x=df.drop('label',axis=1)
```

```python
x
```

```python
y=df.label
```

```python
y
```

```python
from sklearn.model_selection import train_test_split


```

```python
xTrain,xTest,yTrain,yTest=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mx=MinMaxScaler()
xTrain=mx.fit_transform(xTrain)
xTest=mx.fit_transform(xTest)
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model=RandomForestClassifier()
```

```python
model
```

```python
model.fit(xTrain,yTrain)
```

```python
yPred=model.predict(xTest)
```

```python
score=accuracy_score(yPred,yTest)
```

```python
score
```

```python
def recommend(N,P,K,temperature,humidity,ph,rainfall):
    features=np.array([[N,P,K,temperature,humidity,ph,rainfall]])
    mx_features=mx.transform(features)
    label=model.predict(mx_features)[0]
    crop=reverse_crop_dict[label]
    return crop
```

```python
N=45
P=69
K=70
temperature=36
humidity=70
ph=5
rainfall=360

recommend(N,P,K,temperature,humidity,ph,rainfall)

```
